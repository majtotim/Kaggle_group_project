# Predicting the Spread of West Nile Virus in Chicago

The aim of this project is to predict the spread of West Nile Virus in Chicago (see https://www.kaggle.com/c/predict-west-nile-virus).  The data used for this project includes detailed weather measurements and counts of infected vs non-infected mosquitos caught in traps located throughout the Chicago metropolitan area over the course of several years.  Since weather conditions play an important role in the lifecycle of mosquitos, I focused on developing independent variables using aggregated meteorological data covering a short-term and long-term time frames. I utilize several machine learning classifiers, including random forest, gradient boost random forest, SVM, and K nearest neighbors.  The gradient boosting random forest model consistently outperformed other models.  For the sake of saving space, I have omitted code wherein I fine-tuned the functions used to generate weather-related independent variables.

```python
#importing data, spray data for the test set (2008,2010,2012,2014) is not provided. Therefore, spray info is not used for the analysis
train = pd.read_csv('./assets/train.csv')
test =pd.read_csv('./assets/test.csv')
weather = pd.read_csv('./assets/weather.csv')
#spray =pd.read_csv('./assets/spray.csv')
mapdata = np.loadtxt("./assets/mapdata_copyright_openstreetmap_contributors.txt")
```

```python
weather.head(2)
```

```python
### Station 2 contains redundant information, so we delete measurements from this station
weather = weather[weather['Station'] == 1]
weather = weather.drop('Station', axis=1)

### Depth, Water1 and Snowfall don't provide any valuable informaion.  The unique values for these columns are
#Depth ['0']
#Water1 ['M']
#SnowFall ['0.0' '  T' '0.1']
### Let's drop these:
weather = weather.drop(['Depth','Water1','SnowFall'], axis=1)

### moreover, we consider it highly doubtful that sunrise and sunset will have predictive value, so we are
### deleting these columns
weather = weather.drop(['Sunrise','Sunset'], axis=1)

weather.head(2)
```

```python
### what are the data types for each column?
for col in weather.columns:
    print col, type(weather[col][0])
```

```python
# for columns which contain strings, we want to know which can be converted into more useful data types and which 
# should remain as text .  Let's look at the contents of each column which contains strings:
for col in weather.columns:
    if type(weather[col][0]) == str:
        print col, weather[col].unique()
```

```python
# It looks like date needs to be converted into datetime format
# CodeSum needs to be converted into dummy columns containing a 0 or 1 value for 
# each classification code
# Non-numeric symbols need to be removed from WetBulb, StnPressure, SeaLevel, PrecipTotal
# AvgSpeed, Tavg, Depart, WetBulb, Heat, Cool, PrecipTotal, StnPressure, 
# SeaLevel need to be converted to float
```

```python
def to_float(value):
    value = value.strip()
    value = value.replace('M','nan')
    value = value.replace('T','0.0')
    value = float(value)
    return(value)
```

```python
### converting string columns to integers
str_columns = ['AvgSpeed', 'Tavg', 'Depart', 'WetBulb', 'Heat', 'Cool', 'PrecipTotal', 'StnPressure', 'SeaLevel']
for col in str_columns:
    weather[col] = weather[col].apply(lambda x: to_float(x))
```

```python
weather.head(5)
```

```python
### creating new columns for dummies
codes = ['HZ', 'VC', 'FU', 'BC', 'SQ', 'FG+', 'MI', 'TS', 'DZ', 'RA', 'BR', 'FG', 'SN']
for code in codes:
    weather[code] = 0
```

```python
def dummy_codes(x,code):
        if code in x:
            return 1
        else:
            return 0  
```

```python
for code in codes:
    weather[code] = weather['CodeSum'].apply(lambda x: dummy_codes(x,code))
weather.head(2)
```

```python
#dropping original CodeSum column
weather = weather.drop('CodeSum', axis=1)
```

```python
### converting dates from string to datetime object and creating new column
weather['Datetime_Date'] = weather.Date.apply(lambda x: datetime.strptime(x,'%Y-%m-%d'))
```

```python
# wind directions are listed as 0-36 (i.e. 0 deg through 360)
# let's create dummy values for each of the directions
weather['northeast'] = weather['ResultDir'].apply(lambda x: 0 if x > 9 else 1)
weather['southeast'] = weather['ResultDir'].apply(lambda x: 1 if x >= 9 and x < 18 else 0)
weather['southwest'] = weather['ResultDir'].apply(lambda x: 1 if x >= 18 and x < 27 else 0)
weather['northwest'] = weather['ResultDir'].apply(lambda x: 1 if x >= 27 else 0)
weather['north'] = weather['ResultDir'].apply(lambda x: 1 if x > 31.5 or x < 4.5 else 0)
weather['east'] = weather['ResultDir'].apply(lambda x: 1 if x >= 4.5 and x < 13.5 else 0)
weather['south'] = weather['ResultDir'].apply(lambda x: 1 if x >= 13.5 and x < 22.5 else 0)
weather['west'] = weather['ResultDir'].apply(lambda x: 1 if x >= 22.5 and x <= 31.5 else 0)
del weather['ResultDir']
```

```python
# creating rolling average, max, min, etc.
columns_to_sum = ['HZ', 'VC', 'FU', 'BC', 'SQ', 'FG+', 'MI', 'TS', 'DZ', 'RA', 'BR', 'FG', 'SN', 
                  'northeast', 'southeast','southwest','northwest', 'north','south','east','west']
columns_to_max_min_avg = ['Tavg','AvgSpeed','WetBulb','DewPoint','StnPressure','PrecipTotal','Tmin','Cool','Tmax','Heat','ResultSpeed']
```

```python
# setting time window
window1 = 40
window2 = 200

for i in columns_to_sum:
    name = i+str(window1)+'sum'
    weather[name] = weather[i].rolling(window=window1, center=False, min_periods=1, win_type='triang').sum()
    name = i+str(window2)+'sum'
    weather[name] = weather[i].rolling(window=window2, center=False, min_periods=1, win_type='triang').sum()
    del weather[i]
for i in columns_to_max_min_avg:
    name = i+str(window1)+'max'
    weather[name] = weather[i].rolling(window=window1, center=False, min_periods=1).max()
    name = i+str(window1)+'min'
    weather[name] = weather[i].rolling(window=window1, center=False, min_periods=1).min()
    name = i+str(window1)+'mean'
    weather[name] = weather[i].rolling(window=window1, center=False, min_periods=1, win_type='triang').mean()
    name = i+str(window1)+'median'
    weather[name] = weather[i].rolling(window=window1, center=False, min_periods=1).median()
    name = i+str(window2)+'max'
    weather[name] = weather[i].rolling(window=window2, center=False, min_periods=1).max()
    name = i+str(window2)+'min'
    weather[name] = weather[i].rolling(window=window2, center=False, min_periods=1).min()
    name = i+str(window2)+'mean'
    weather[name] = weather[i].rolling(window=window2, center=False, min_periods=1, win_type='triang').mean()
    name = i+str(window2)+'median'
    weather[name] = weather[i].rolling(window=window2, center=False, min_periods=1).median()
    del weather[i]
```

## Models

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import Lasso
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.metrics import roc_auc_score,roc_curve, auc
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
```

```python
# model evaluation function
def evaluate_model(model):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    a = accuracy_score(y_test, y_pred)
    probabilities = model.predict_proba(X_test)
    #cm = confusion_matrix(y_test, y_pred)
    conmat = np.array(confusion_matrix(y_test, y_pred, labels=[1,0]))

    confusion = pd.DataFrame(conmat, index=['1', '0'],
                         columns=['predicted_1','predicted_0'])

    cr = classification_report(y_test, y_pred)
    
    print 'confusion matrix:'
    print confusion
    print 'classification_report:'
    print cr
    print 'Accuracy of the model on test:',a
    return probabilities

# plotting the ROC_AUC curve
def ROC_AUC_DRAW(model):
    probabilities = model.best_estimator_.predict_proba(X_test)
    y_score=[]
    for i in range(0,len(probabilities)):
        y_score.append(probabilities[i][1])


    FPR = dict()
    TPR = dict()
    ROC_AUC = dict()

    # For class 1, find the area under the curve
    FPR[1], TPR[1], _ = roc_curve(y_test, y_score)
    ROC_AUC[1] = auc(FPR[1], TPR[1])

    # Plot of a ROC curve for class 1 
    plt.figure(figsize=[5,5])
    plt.plot(FPR[1], TPR[1], label='ROC curve (area = %0.2f)' % ROC_AUC[1], linewidth=4)
    plt.plot([0, 1], [0, 1], 'k--', linewidth=4)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate', fontsize=12)
    plt.ylabel('True Positive Rate', fontsize=12)
    plt.title('Receiver operating characteristic for predicting WNV', fontsize=12)
    plt.legend(loc="lower right")
    plt.savefig('ROC_AUC.png')
    plt.show()
```

## Random Forest

```python
##Grid search over Random Forest parameters

max_depths = [0.3,0.5,0.7,1.0]
max_features = [0.3,0.5,0.7,1.0]
n_estimators = [10,100,300]

rf = RandomForestClassifier(n_jobs=-1,random_state = 33, class_weight='balanced')
gsrf = GridSearchCV(estimator = rf,param_grid=dict(max_depth = max_depths, max_features=max_features,n_estimators=n_estimators), n_jobs=-1,cv=3,scoring='roc_auc')
gsrf.fit(X_train, y_train)
print 'best parameters for the model:',gsrf.best_params_
print 'best score on train:',gsrf.best_score_
probability = evaluate_model(gsrf.best_estimator_)
```

```python
# Getting the probabilities and save to csv file
probabilities_rf = gsrf.best_estimator_.predict_proba(test_add)
pro_rf =probabilities_rf[:,1]
pro_rf = pd.Series(pro_rf)
submit= pd.concat([test['Id'], pro_rf], axis= 1)
submit.columns=['Id','WnvPresent']
submit.to_csv('gsrf2_pro.csv',index = False)
```
